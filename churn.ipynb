STUDENT PRODUCTIVITY ANALYSIS USING PYTHON (PANDAS)

```python
import pandas as pd

df = pd.read_csv("student_data.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
df.groupby("Study Hours")["Productivity Score"].mean()
```

```python
df.groupby("Sleep Hours")["Productivity Score"].mean()
```

```python
df.groupby("Screen Time")["Productivity Score"].mean()
```

```python
df.groupby("Exercise")["Productivity Score"].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="Study Hours", y="Productivity Score", data=df)
plt.title("Study Hours vs Productivity")
plt.show()
```

Study vs Productivity
    -->This graph shows that students who study more tend to have higher productivity. The trend is not perfect, but overall there is a clear positive relationship.

```python
sns.scatterplot(x="Sleep Hours", y="Productivity Score", data=df)
plt.title("Sleep vs Productivity")
plt.show()
```

Sleep vs Productivity
   -->From this graph, it can be seen that moderate sleep (around 6–8 hours) is linked with better productivity. Too little sleep seems to reduce performance.

```python
sns.scatterplot(x="Screen Time", y="Productivity Score", data=df)
plt.title("Screen Time vs Productivity")
plt.show()
```

Screen Time vs Productivity
   -->This graph shows that as screen time increases, productivity tends to decrease. This could be due to distractions affecting focus.

```python
sns.boxplot(x="Exercise", y="Productivity Score", data=df)
plt.title("Exercise vs Productivity")
plt.show()
```

Exercise vs Productivity:
   -->students who exercise generally have better productivity scores. This suggests that physical activity may help improve focus and performance.

KEY INSIGHTS:
   1. Students who study more generally perform better, which shows that consistent study habits improve productivity.
   2.Getting around 6–8 hours of sleep seems to be ideal, as students with very low sleep tend to perform worse.
   3.High screen time clearly affects productivity in a negative way, probably due to distractions like social media.
   4.Students who exercise regularly appear to be more productive compared to those who do not, showing that physical activity also plays a role.

CONCLUSION:
   -->Overall, the analysis suggests that maintaining a balanced lifestyle is important for better productivity. Managing study time, getting enough sleep, reducing screen time, and staying physically active can all help improve performance.
